```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df_file = "city_temperature.csv"
```

```python
df = pd.read_csv(df_file)
df.head()
```

```python
usdf = df[df["Country"].str.contains('US')]
```

```python
usdf.head()
```

```python
usdf = usdf.loc[(usdf["Year"] > 2019) & (usdf["Year"] < 2021)]
```
